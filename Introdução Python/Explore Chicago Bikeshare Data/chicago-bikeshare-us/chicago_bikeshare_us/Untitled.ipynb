```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('chicago.csv')
```

```python
df[:20]
```

```python
df.Gender[:20]
```

```python
def column2list(Column):
    return list(Column)
```

```python
pd.value_counts(df.Gender)
```

```python
def count_genders(gender):
    return pd.value_counts(df.Gender[df.Gender == gender]).values[0]
```

```python
count_genders('Male')
```

```python
labels = list(pd.value_counts(df.Gender).index)
values = list(pd.value_counts(df.Gender).values)
plt.pie(values,labels=labels, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```
